```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sb_df = pd.read_csv("superbowl.csv")
```

```python
sb_df.head(5)
```

```python
sb_df.info()
```

```python
sb_df.describe()
```

```python
sns.set(style="whitegrid")
```

```python
graph = sns.jointplot("Winner Pts", "Loser Pts", data=sb_df, kind="reg",
                  xlim=(0, 60), ylim=(0, 60), color="m", height=10)
plt.show(graph)
```

Story: When comparing the points the losing team score in superbowl past, compared to the winnning teams score, you can see that it is ususally a close game, with few outliers

```python
broadband_df = pd.read_csv('broadband-speed-test.csv')
```

```python
broadband_df.head(5)
```

```python
sns.set(style="whitegrid")
```

```python
graph = sns.jointplot("advertised_download", "actual_download", data=broadband_df, kind="reg",
                  xlim=(0, 100), ylim=(0, 100), color="m", height=10)
plt.show(graph)
```

story: broadband internet is pretty inconsistent on what download speed you get from them comparted to what the advertised speed is
